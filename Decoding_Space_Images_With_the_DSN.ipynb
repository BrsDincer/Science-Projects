# Background

[NASA’s Deep Space Network](https://www.jpl.nasa.gov/missions/dsn), or **DSN**, is a global network of antennae that we use to communicate with faraway spacecraft. With stations roughly every 120 degrees around the globe – near Goldstone, California; Canberra, Australia; and Madrid, Spain – the DSN provides a constant link to robotic explorers venturing beyond Earth’s Moon and to some Earth-orbiting missions.

Data transmitted by a spacecraft via the DSN are encoded into `binary code` for transmission and then decoded on Earth.

- Binary is a base-2 number system that uses two states, 1 and 0, to represent a number. A bit is a single binary digit that is read by computers as on (1) or off (0). The number 10110100 is a sequence of 8 bits and is read as individual digits – one zero one one zero one zero zero – to distinguish from how we read decimal numbers.
- Standard notation for numerals in different base systems includes a subscript indicating the base, so sometimes a binary number will appear with a subscript of 2, such as $10110100_{2}$. The subscript for decimal numbers is often omitted unless it is needed for context.
- In base 2, we can use the numerals 0-1 in any place value to represent a quantity. For example, the binary number $1101 = (1 x 23) + (1 x 22) + (0 x 21) + (1 x 20) = 8 + 4 + 0 + 1 = 13$ (in decimal, or base 10, notation). Examples:

$10002 = 8_{10}$

$1012 = 2^{2} + 2^{0} = 5_{10}$

$1100112 = 2^{5} + 2^{4} + 2^{1} + 2^{0} = 51_{10}$

- Computers use billions of bit sequences to store complex information. These groups of 1s and 0s might not look like anything; however, in binary code, they translate to a message or image.

The reason that spacecraft and the DSN use binary code to communicate is that it's easier for computers to process than decimal notation or text, and it also takes up less memory. Binary transmissions can also better withstand errors from signal noise caused by interference. If a noisy transmission is received, as long as 1 and 0 can be differentiated, the message can be deciphered, cleaned, and restored to a pristine state of 1s and 0s.

**Bitmap images** are representations in which each item corresponds to one or more bits of information. Bitmap image files can store large images by breaking the image down into individual dots, called pixels, and assigning a binary number to describe how bright each pixel is. When a spacecraft camera captures an image, light passes through the camera’s filters to a detector chip, where the brightness of each pixel is recorded for each color filter (typically red, green, and blue, or RGB) to produce a color code for each pixel of the image. Separate bitmap images for each color are transmitted to Earth, converted back into RGB color codes, and then combined to create a single full-color image. In addition to colors, other filters can be used to pick out particular spectral lines that are of scientific interest.

## Error-Module

```python
class ErrorModule(object):
  def __init__(self)->CLASSINIT:
    self.error = NotImplementedError(NotImplemented)
  def __str__(self)->str:
    return "Error Module - Sub(Script)"
  def __call__(self)->ERROR:
    return self.error
  def __getstate__(self)->ERROR:
    raise self.error
  def __repr__(self)->DOCUMENTATION:
    return ErrorModule.__doc__
  @property
  def Default(self)->ERROR:
    raise self.error
  def Manuel(self,errorType:ERROR,errorMessage:str)->ERROR:
    raise errorType(errorMessage)
```

## Process

```python
class TransformationProcess(object):
  def __init__(self,colorMap:None=None)->CLASSINIT:
    if colorMap is None:
      self.mapping = ColorMapping
    else:
      self.mapping = colorMap
  def __str__(self)->str:
    return "Transformation Process - Main Script"
  def __call__(self)->NULL|None:
    return None
  def __repr__(self)->DOCUMENTATION|str:
    return TransformationProcess.__doc__
  def __getstate__(self)->ERROR:
    ErrorModule().Default
  def CompatibleFormat(self,imageArray:np.ndarray|list)->PROCESS:
    colorGrid = []
    for row in imageArray:
      colorRow = []
      for pixel in row:
        closest = self.FindClosestColor(color=pixel)
        colorRow.append(self.mapping[closest])
      colorGrid.append(colorRow)
    return colorGrid
  def BinaryGridToColor(self,binaryGrid:DATAGRID)->PROCESS:
    colorGrid = [[self.mapping[binary] for binary in row] for row in binaryGrid]
    return colorGrid
  def SaveColorGrid(self,colorGrid:DATAGRID,fileName:None=None)->PROCESS:
    CreateDirectory("image_result")
    if fileName is not None:
      savePath = os.path.join(os.getcwd(),f"image_result/{fileName}.jpeg")
    else:
      savePath = os.path.join(os.getcwd(),"image_result/image_grid.jpeg")
    array = np.array(colorGrid,dtype=np.uint8)
    try:
      image = Image.fromarray(array,"RGB")
    except Exception as err:
      print(f"ERROR: {err}")
      image = self.CompatibleFormat(array)
      image = Image.fromarray(image,"RGB")
    image.save(savePath)
    return savePath
  def FindClosestColor(self,color:int)->PROCESS:
    mappingValues = np.array(list(self.mapping.values()))
    colorDiff = mappingValues - np.array(color)
    distance = np.sum(colorDiff**2,axis=1)
    smallest = np.argmin(distance)
    colosestIndex = list(self.mapping.keys())[smallest]
    closestIndex = np.argmin(cdist([color],mappingValues,metric="euclidean"))
    return list(self.mapping.keys())[smallest]
  def ImageToBinaryGrid(self,imagePath:str)->PROCESS:
    image = Image.open(imagePath).convert("RGB")
    array = np.array(image,dtype=np.uint8)
    grid = []
    for row in array:
      binaryRow = []
      for pixel in row:
        closestColor = self.FindClosestColor(pixel)
        binaryRow.append(closestColor)
      grid.append(binaryRow)
    return grid
```

## Binary Grid Creation

> Create a patterned binary grid of a given size with specified patterns

```python
def CreatePatternedBinary(size:int,patterns:list,colorMap:dict=ColorMapping)->PROCESS:
  grid = []
  quadrant = size//len(patterns)
  for index in range(size):
    row = []
    for subIndex in range(size):
      quadrantX = index//quadrant
      quadrantY = subIndex//quadrant
      patternIndex = quadrantX*len(patterns)+quadrantY
      binaryCode = patterns[patternIndex%len(patterns)]
      if binaryCode not in colorMap:
        ErrorModule().Manuel(ValueError,f"{binaryCode} - NOT FOUND")
      row.append(binaryCode)
    grid.append(row)
  return grid
```

## Color Map Creation

> This function creates a linear distribution of colors across the RGB color space and assigns each one a unique 7-bit binary code. The generate_color_mapping function can be called to get a dictionary where the keys are 7-bit binary strings and the values are RGB color tuples.

```python
def GenerateColorMapping(numberColor:int=400)->PROCESS:
  colors = []
  step = int(255*255*255/numberColor)
  for index in range(0,255*255*255,step):
    blue = index%255
    green = (index//255)%255
    red = (index//(255*255))%255
    colors.append((red,green,blue))
  colorMap = {}
  for index,color in enumerate(colors):
    binary = format(index,"05b")
    colorMap[binary] = color
  return colorMap
```

```python
def FindClosestColor(pixelValue:int|tuple,colorMapInitial:dict)->PROCESS:
  closest = sorted(colorMapInitial.items(),key=lambda item: ColorDistance(pixelValue,item[1]))
  return closest[0][0]
```

```python
def GetColorMapAndOutputImage(imageMain:IMAGE,colorMapInitial:dict)->PROCESS:
  binaryCustomData = []
  if imageMain.mode != "RGB":
    imageMain = imageMain.convert("RGB")
  for y in range(imageMain.size[1]):
    for x in range(imageMain.size[0]):
      pixel = imageMain.getpixel((x,y))
      binaryCustomData.append(FindClosestColor(pixelValue=pixel,colorMapInitial=colorMapInitial))
  decodedImageCustomData = [colorMapInitial[binary] for binary in binaryCustomData]
  decodedNewImage = Image.new("RGB",imageMain.size)
  decodedNewImage.putdata(decodedImageCustomData)
  CreateDirectory("new_image_result")
  CreateDirectory("new_image_binary")
  pathImage = os.path.join(os.getcwd(),"new_image_result/decoded_image.jpg")
  pathBinary = os.path.join(os.getcwd(),"new_image_binary/decoded_binary.txt")
  with open(pathBinary,"w") as fileOps:
    fileOps.write(",".join(binaryCustomData))
  print(f"SAVED BINARY IMAGE FORMAT ON: {pathBinary}")
  decodedNewImage.save(pathImage)
  print(f"SAVED IMAGE ON: {pathImage}")
  return decodedNewImage
```
